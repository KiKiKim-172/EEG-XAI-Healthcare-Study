```python
# MNE sample EEG data (built-in stimulus example)
data = np.random.randn(1, 500)  # 1 channel, 500 samples
info = mne.create_info(ch_names=['Fz'], sfreq=250, ch_types=['eeg'])
raw = mne.io.RawArray(data, info)

# Plot
raw.plot(title="Test EEG Plot")
```
